```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

numb_schools = len(school_data_complete["school_name"].unique())
numb_students = school_data_complete["Student ID"].max()+1
school_budget = school_data_complete.groupby(["school_name", "budget"]).size()\
                .reset_index().rename(columns={0:"Students"})
student_scores = school_data_complete.groupby(["Student ID", "reading_score", "math_score"]).size()\
                .reset_index().rename(columns={0:"Students"})
student_scores["avg_score"] = (student_scores["reading_score"] + student_scores["math_score"])/2
#total_budget = school_budget["budget"]
#school_data_complete.head()
pre_results_df = pd.DataFrame([{
    "Total Schools":numb_schools,
    "Total Students":"{:,}".format(numb_students),
    "Total Budget": "${:20,.2f}".format(school_budget["budget"].sum()),
    "Avg Reading Score" : "{0:.2%}".format(student_scores["reading_score"].mean()/100),
    "Avg Math Score" : "{0:.2%}".format(student_scores["math_score"].mean()/100),
    "% Passing Reading" : "{0:.2%}".format(len(student_scores[(student_scores['reading_score']>=70)])/numb_students),
    "% Passing Math" : "{0:.2%}".format(len(student_scores[(student_scores['math_score']>=70)])/numb_students),
    "% of all tests passed" : "{0:.2%}".format(student_scores["avg_score"].mean()/100)
}])
#Format and reorder
results_df=pre_results_df[["Total Schools", "Total Students", "Total Budget", "Avg Math Score", 
                           "Avg Reading Score", "% Passing Math", "% Passing Reading", "% of all tests passed"]]

results_df
```

```python
def school_summaries(x):
    d={}
    d["student_count"]=x["Student ID"].count()
    d["student_count_fmt"]="{:,}".format(d["student_count"])
    d["budget"] = x["budget"].max()
    d["budget_fmt"] = "${:0}".format(d["budget"])
    d["per_student_budget"] = d["budget"]/d["student_count"]
    d["per_student_budget_fmt"] = "${:.2f}".format(d["per_student_budget"])
    d["avg_reading"] = x["reading_score"].mean()

    d["avg_math"] = x["math_score"].mean()

    d["per_pass_reading"] = len(x[(x['reading_score']>=70)])/d["student_count"]
    d["per_pass_reading_fmt"] = "{:.2%}".format(d["per_pass_reading"])
    d["per_pass_math"] = len(x[(x['math_score']>=70)])/d["student_count"]
    d["per_pass_math_fmt"] = "{:.2%}".format(d["per_pass_math"])
    d["overall_pass_rate"] = (d["per_pass_reading"] + d["per_pass_math"])/2
    d["overall_pass_rate_fmt"] = "{:.3%}".format(d["overall_pass_rate"])
    return pd.Series(d, index=[    "student_count", "student_count_fmt", "budget", "budget_fmt",
                                "per_student_budget", "per_student_budget_fmt", "avg_reading",
                                "avg_math", "per_pass_reading", "per_pass_reading_fmt", "per_pass_math",
                                "per_pass_math_fmt", "overall_pass_rate", "overall_pass_rate_fmt"])

school_sums = pd.DataFrame(school_data_complete.groupby(["school_name", "type"], as_index=False).apply(school_summaries))
schools_sums = school_sums.reset_index()
school_sums_formatted = school_sums[["student_count_fmt", "budget_fmt", "per_student_budget_fmt", "avg_math", "avg_reading",
                                     "per_pass_math_fmt", "per_pass_reading_fmt", "overall_pass_rate_fmt" ]]
school_sums_formatted.sort_values("overall_pass_rate_fmt", ascending=False).head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
scores_9th = school_data_complete[["school_name", "math_score"]].loc[school_data_complete["grade"]=="9th"]\
            .groupby("school_name").agg({"math_score":"mean"})\
            .rename(columns={"math_score":"math_score_9th"})
scores_10th = school_data_complete[["school_name", "math_score"]].loc[school_data_complete["grade"]=="10th"]\
            .groupby("school_name").agg({"math_score":"mean"})\
            .rename(columns={"math_score":"math_score_10th"})
scores_11th = school_data_complete[["school_name", "math_score"]].loc[school_data_complete["grade"]=="11th"]\
            .groupby("school_name").agg({"math_score":"mean"})\
            .rename(columns={"math_score":"math_score_11th"})
scores_12th = school_data_complete[["school_name", "math_score"]].loc[school_data_complete["grade"]=="12th"]\
            .groupby("school_name").agg({"math_score":"mean"})\
            .rename(columns={"math_score":"math_score_12th"})

math_score_sums = pd.merge(pd.merge(pd.merge(scores_9th, scores_10th, on = "school_name"),scores_11th, on = "school_name")
                ,scores_12th, on = "school_name")
math_score_sums
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
scores_9th = school_data_complete[["school_name", "reading_score"]].loc[school_data_complete["grade"]=="9th"]\
            .groupby("school_name").agg({"reading_score":"mean"})\
            .rename(columns={"reading_score":"reading_score_9th"})
scores_10th = school_data_complete[["school_name", "reading_score"]].loc[school_data_complete["grade"]=="10th"]\
            .groupby("school_name").agg({"reading_score":"mean"})\
            .rename(columns={"reading_score":"reading_score_10th"})
scores_11th = school_data_complete[["school_name", "reading_score"]].loc[school_data_complete["grade"]=="11th"]\
            .groupby("school_name").agg({"reading_score":"mean"})\
            .rename(columns={"reading_score":"reading_score_11th"})
scores_12th = school_data_complete[["school_name", "reading_score"]].loc[school_data_complete["grade"]=="12th"]\
            .groupby("school_name").agg({"reading_score":"mean"})\
            .rename(columns={"reading_score":"reading_score_12th"})

reading_score_sums = pd.merge(pd.merge(pd.merge(scores_9th, scores_10th, on = "school_name"),scores_11th, on = "school_name")
                ,scores_12th, on = "school_name")
reading_score_sums
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.

spending_bins = [school_sums.per_student_budget.min(), school_sums.per_student_budget.quantile(.25), 
                 school_sums.per_student_budget.quantile(.50)
                 , school_sums.per_student_budget.quantile(.75), school_sums.per_student_budget.max()]
group_names = ["1st Quartile", "2nd Quartile", "3rd Quartile", "4th Quartile"]
school_sums["spending_bin"] = pd.cut(school_sums["per_student_budget"], bins=spending_bins, labels=group_names, include_lowest = True)


#merged back w/ complete student data to get new average
school_data_complete2 = pd.merge(school_data_complete, school_sums, on="school_name")

school_data_complete2.groupby("spending_bin")[["avg_math", "avg_reading", "per_pass_math", "per_pass_reading", "overall_pass_rate"]].mean()
```
